```python
df_12600 = df_komegasst[df_komegasst["Case"].str.contains('CNDV_12600')==True]
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(311)
ax.scatter(df_12600['komegasst_Cx'],
                df_12600['komegasst_Cy'],
                c=df_12600['komegasst_Ux'],s=3)
ax.set_title(r'$U_x$')
ax.set_aspect(1)

ax = fig.add_subplot(312)
ax.scatter(df_12600['komegasst_Cx'],
                df_12600['komegasst_Cy'],
                c=df_12600['komegasst_wallDistance'],s=3)
ax.set_title('Wall distance')

ax.set_aspect(1)

ax = fig.add_subplot(313)
ax.scatter(df_12600['komegasst_Cx'],
                df_12600['komegasst_Cy'],
                c=df_12600['labels_k'],s=3)
ax.set_title('Label field (in this case, DNS): turbulent kinetic energy')

ax.set_aspect(1)
fig.tight_layout()
```
